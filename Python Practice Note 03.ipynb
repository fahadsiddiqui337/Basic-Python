# 3. Dictionary 
Dictionaries are ordered collections of unique values stored in (Key-Value) pairs.





```python
d1 = {'Apple':50, 'Mango':70, 'Guava':90, 'Banana':110} # here like Apple is key and Value is 50 and so on
d1
```

```python
type(d1)
```

# Extracting Keys and Values


```python
# Extracting keys

Fruit = {'Apple':50, 'Mango':70, 'Guava':90, 'Banana':110}
Fruit.keys()
```

```python
# Extracting Value

Fruit = {'Apple':50, 'Mango':70, 'Guava':90, 'Banana':110}
Fruit.values()

```

# Modifying 

```python
# Adding a new Elements

Fruit = {'Apple':50,'Guava':90, 'Banana':110}
Fruit["Mango"] = 50
Fruit
```

```python
# Changing an Existing Element

Fruit = {'Apple':50,'Guava':90, 'Banana':110, 'Mango':30}
Fruit["Mango"] = 50
Fruit

```

# Dictionary Functions



```python
# Update one dictionary with another 

Fruit1 = {"Apple":10, "Orange":30}
Fruit2 = {"Banana":30, "Guava":80 }
Fruit1.update(Fruit2)

Fruit1

```

```python
# Popping Elements

Fruit = {'Apple':50,'Guava':90, 'Banana':110, 'Mango':30}
Fruit.pop("Apple")
Fruit
```

# 4.Set

In Python, a Set is an unordered collection of data items that are unique. In other words, Python Set is a collection of elements (Or objects) that contains no duplicate elements





```python
s1 = {1, 3.4, "Fahad"}
s1      
```

```python
# Set Does Not Allow Duplicates
s1 = {1, 3.4, "Fahad",3.4,"Fahad"}
s1
```

# Set Operation

```python
# Update one dictionary's elements with another

s1 = {1, "a", True, 2, "b", False}
s1.add("Hello")
s1
```

```python
# Updating multiple elements

s1 = {1, "a", True, 2, "b", False}
s1.update([10,20,30])
s1
```

```python
# Removing an element

s1 = {1, "a", True, 2, "b", False}
s1.remove("b")
s1
```

# Set Functions

```python
# Union of Two sets

s1 = {1,2,3}
s2 = {"a","b","c"}
s1.union(s2)
```

```python
# Intersection of Two Sets

s1 = {1,2,3,4,5,6}
s2 = {5,6,7,4,8,9,0}
s1.intersection(s2)
```

# Python Conditions and If statements

Python supports the usual logical conditions from mathematics:

 Equals: a == b
 
 Not Equals: a != b
 
 Less than: a < b
 
 Less than or equal to: a <= b
 
 Greater than: a > b
 
 Greater than or equal to: a >= b




These conditions can be used in several ways, most commonly in "if statements" and loops.

An "if statement" is written by using the if keyword
    

```python
a = 10
b = 20

if a>b:
    print("B is greater then A")
else:
    print("A is greater than B")
```

```python
a = 33
b = 33
if a > b:
    print("B is greater then a")
elif a == b:
    print("A is equal to B")
```

# Nested If
You can have if statements inside if statements, this is called nested if statements."if else" statements in Python and how they can be used to test multiple conditions. In Python programming, sometimes you need to check multiple conditions before making a decision. That's where nested if-else statements come in.

```python
# Examples:01

x = 41
if x > 10:
    print("Above ten,")
    if x > 20:
        print("and also above 20!")
    else: 
        print("but not above 20.")

```

```python
num = 5
if num > 0:
    print("The number is positive")
else:
    if num < 0:
        print("The Number is negative.")
    else:
        print("The Number is zero.")
```

```python
 age = 35

if age >= 60:
    print("You are a senior citizen")
else:
    if age >= 18:
        print("You are an adult.")
    else:
        print("You are teenager.")
```
