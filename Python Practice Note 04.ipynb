# Loop in python
In computer programming, loops are used to repeat a block of code.

For example, if we want to show a message 100 times, then we can use a loop. It's just a simple example; you can achieve much more with loops.

There are 2 types of loops in Python:

for loop

while loop



# While Loop



```python
# Write a Code with can generate COunting 1 to 10

i = 1
while i<=10:
    print(i)
    i =i+1
```

```python
# Print i as long as i is less than 6:
i = 1
while i<6:
    print(i)
    i = i+1
```

```python
# Break Statement: Exit the loop whin i is 3

i = 1
while i < 6:
  print(i)
  if (i == 3):
    break     
  i += 1
```

```python
# The Continue Statement: Continue to the text iteration if i is 3

i = 0
while i < 6:
    i = i+1
    if i == 3:
        continue
    print(i) 
```

```python
# Write a code which generate Table of 2

i = 1
n = 2
while i<=10:
    print(n," * ",i," = ",n*i)
    i=i+1
```

# For Loop


```python
# list in For loop

l1 = ['mango','apple','grapes','orange']
for i in l1:
    print(i)
```

```python
# looping through a string

for i in "banana":
    print(i)
```

```python
# The Break Statment

fruits = ["apple","banana","cherry"]
for i in fruits:
    print(i)
    if i== "banana":
        break
        print(i)
```

```python
# The Continue Statment

fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        continue
    print(x)
        
```

```python
# for loop in Multiple list or Nested Loop

l1 = ['orange','black','white']
l2 = ['chair','book','laptop']
for i in l1:
    for j in l2:
        print(i,j)
```

```python
# Print First 10 numbers using for loop

for num in range(10):
    print(num)
```

```python
# Print sum of all even Numbers from 10 to 20

sum = 0
for i in range(2,22,2):
    sum = sum+i
    print(sum)
```

```python
#Calculate the square of eac numbers of list

numbers = [1,2,3,4,5]
for i in numbers:
    square = i**2
    print("Square of:",i,"is",square)
    
```
